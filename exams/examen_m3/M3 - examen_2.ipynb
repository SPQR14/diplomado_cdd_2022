```python
from re import sub
from decimal import Decimal

def money_formartting(amount):
    return Decimal(sub(r'[^\d.]', '', amount))
```

```python
df['PerCapitaIncome-Zipcode'] = df['PerCapitaIncome-Zipcode'].map(money_formartting)
df['YearlyIncome-Person'] = df['YearlyIncome-Person'].map(money_formartting)
df['TotalDebt'] = df['TotalDebt'].map(money_formartting)
df['CreditLimit'] = df['CreditLimit'].map(money_formartting)
df['Amount'] = df['Amount'].map(money_formartting)
```
