```python
import numpy as np
 
def PCA(X , num_components):
     
    #1
    X_meaned = X - np.mean(X , axis = 0)
     
    #2
    cov_mat = np.cov(X_meaned , rowvar = False)
     
    #3
    eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
     
    #4
    sorted_index = np.argsort(eigen_values)[::-1]
    sorted_eigenvalue = eigen_values[sorted_index]
    sorted_eigenvectors = eigen_vectors[:,sorted_index]
     
    #5
    eigenvector_subset = sorted_eigenvectors[:,0:num_components]
     
    #6
    X_reduced = np.dot(eigenvector_subset.transpose() , X_meaned.transpose() ).transpose()
     
    return X_reduced
```

## VARClus

- El algoritmo general es
- 1. Realiza PCA en el espacio de funciones actual. Si el valor propio máximo es menor que el umbral, detener la descomposición
- 2. Calcula los primeros componentes N PCA y asigna características a estos componentes según correlación absoluta de mayor a menor. Estos componentes son los centroides iniciales de estos grupos de variables.
- 3. Después de la asignación inicial, el algoritmo realiza una asignación iterativa denominada Más cercano Clasificación de componentes (NCS). Básicamente, los vectores centroides se vuelven a calcular como los primeros componentes de los grupos secundarios y el algoritmo reasignará cada una de las características basado en la misma regla de correlación.
- 4. Después de NCS, el algoritmo intenta aumentar la varianza total explicada por la primera
            Componente PCA de cada clúster secundario mediante la reasignación de funciones entre clústeres
