# <h1 style="color:#3b4879"> Codificación a nivel nominal </h1>  

Codificación categórica
Una forma de mejorar sus predicciones es aplicar métodos inteligentes al trabajar con variables categóricas. Estas variables, como su nombre indica, tienen valores discretos y representan algún tipo de categoría o clase. Por ejemplo, el color puede ser una variable categórica ("rojo", "azul", "verde"). El desafío es incluir estas variables en el análisis de datos y usarlas con algoritmos de aprendizaje automático. Algunos algoritmos de aprendizaje automático admiten variables categóricas sin más manipulación, pero otros no. Por eso utilizamos una codificación categórica.

## <h1 style="color:#a56d5d"> One-Hot Encoding </h1>   

Esta es una de las técnicas de codificación categórica más populares. Distribuye los valores de una entidad a varias entidades de marca y les asigna valores 0 o 1. Este valor binario representa la relación entre características codificadas y no codificadas.

El método principal que tenemos es transformar nuestro datos categóricos en variables "dummy". 

Las variables "dummy" toman el valor cero o uno para indicar la ausencia o presencia de un categoría. 
Son suplentes numéricos, para datos cualitativos.

Ejemplo: Género : "hombre","mujer", en este caso usaríamos un código ficticio cuando la persona es  mujer = 1,
y hombre = 0 cuando la persona es hombre.

Al trabajar con variables "dummy", es importante evitar información duplicada.

## <h1 style="color:#a56d5d"> TARGET ENCODING </h1>   

- Reemplaza un valor categórico con el valor promedio de la salida (es decir, el objetivo) para ese valor de la característica. Básicamente, todo lo que necesita hacer es calcular la salida promedio para todas las filas con un valor de categoría específico. Ahora bien, esto es bastante sencillo cuando el valor de salida es numérico.
- Es una codificación es la cuál es necesario realizar un conjunto de entrenamiento y uno de prueba ya que para que se pueda medir el desempeño de un modelo necesitamos un conjunto de registros no antes vistos, por lo cúal para target encoding los valores promedios por categoria deben obtenerse unicamente del conjunto de entrenamiento , excluyendo los valores de prueba.

Clasificación : https://towardsdatascience.com/all-about-target-encoding-d356c4e9e82

```python
##AGGREGATION FUNCTIONS
def minPctChange(x):
    return x.pct_change().replace([np.inf, -np.inf], np.nan).min()
def maxPctChange(x):
    return x.pct_change().replace([np.inf, -np.inf], np.nan).max()
def meanPctChange(x):
    return x.pct_change().replace([np.inf, -np.inf], np.nan).mean()
def stdPctChange(x):
    return x.pct_change().replace([np.inf, -np.inf], np.nan).std()
def sumPctChange(x):
    return x.pct_change().replace([np.inf, -np.inf], np.nan).sum()
def minChange(x):
    return x.diff().replace([np.inf, -np.inf], np.nan).min()
def maxChange(x):
    return x.diff().replace([np.inf, -np.inf], np.nan).max()
def meanChange(x):
    return x.diff().replace([np.inf, -np.inf], np.nan).mean()
def stdChange(x):
    return x.diff().replace([np.inf, -np.inf], np.nan).std()
def sumChange(x):
    return x.diff().replace([np.inf, -np.inf], np.nan).sum()
```
